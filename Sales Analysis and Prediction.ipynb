```python
date_data = features.groupby("Date").agg({"Temperature":"mean"
                                            ,"Fuel_Price":"mean"
                                            ,"IsHoliday":"sum"
                                            ,"CPI":"mean"
                                           ,"Unemployment":"mean"})
date_data = date_data.sort_index()
```

```python
date_sales_data = sales.groupby("Date").agg({"Weekly_Sales":"sum"})
date_sales_data.sort_index(inplace=True)
date_sales_data.Weekly_Sales = date_sales_data.Weekly_Sales/1000000
date_sales_data.Weekly_Sales = date_sales_data.Weekly_Sales.apply(int)
data = pd.merge(date_sales_data, date_data, left_index=True,right_index=True, how='left')
```

```python
data["IsHoliday"] = data["IsHoliday"].apply(lambda x: True if x == 45.0 else False )
```

```python
data['Temperature'] = (data['Temperature']- 32) * 5./9.
data.head()
```

```python
fig = plt.figure()

fig, ax = plt.subplots(5,1,figsize=(12,10),sharex=True) 

data["Weekly_Sales"].plot(ax=ax[0],title="Weekly Sales / Sales on Holiday", color='Black')

data[data.IsHoliday==True]["Weekly_Sales"].plot(marker="o",ax=ax[0], legend = 'Holiday week sales', color='Orange')

data["Temperature"].plot(ax=ax[1], title="Temperature",color='Black')

data["Fuel_Price"].plot(ax=ax[2],title="Fuel_Price", color='Black')

data["CPI"].plot(ax=ax[3],title="CPI", color='Black')

data["Unemployment"].plot(ax=ax[4],title="Unemployment", color='Black')
```
